```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sem = [1,2,3,4,5]
spi=[7.8,8.9,9.7,8.5,9.8]
cpi=[7.2,8.2,9.2,9.3,9.2]
plt.plot(sem,spi)
```

```python
plt.figure(figsize=(15,10))
sem = [1,2,3,4,5]
spi=[7.8,8.9,9.7,8.5,9.8]
cpi=[7.2,8.2,9.2,9.3,9.2]
plt.plot(sem,spi,'*',linewidth=3,linestyle='dotted',markersize=18)
plt.plot(sem,cpi,'*',markersize=12,color='red')
plt.legend(["spi","cpi"])
plt.xticks(sem)

plt.xlabel("semester")
plt.ylabel("range of SPI/CPI", fontsize="10")
plt.title("SPI/CPI vs semester",fontsize="18")
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(sem,spi,c='r',s=100,alpha=0.9)
plt.scatter(sem,cpi,c='b',alpha=0.5)
plt.legend(["spi","cpi"])
plt.xlabel("semester")
plt.ylabel("range of SPI/CPI", fontsize="10")
plt.title("SPI/CPI vs semester",fontsize="18")
plt.show()
```

```python
plt.figure(figsize=(10,7))
age=[2,3,5,4,1,5,7,89,65,23,23,12,34,45,76,34,22,45]
plt.hist(age,bins=10)
plt.legend(["spi","cpi"])
plt.xlabel("age")
plt.ylabel("frequency", fontsize="10")
plt.title("age vs frequency",fontsize="18")
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.bar(sem,spi,color='r')
plt.legend(["spi","cpi"])
plt.xlabel("sem")
plt.ylabel("SPI", fontsize="10")
plt.title("SEM vs SPI",fontsize="18")
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.barh(sem,spi)
plt.legend(["spi","cpi"])
plt.xlabel("sem")
plt.ylabel("SPI", fontsize="10")
plt.title("SEM vs SPI",fontsize="18")
plt.show()
```
