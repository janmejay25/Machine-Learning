<h2>Total number of IPL Seasons</h2>

```python
total_seasons=df['Season'].unique().shape
total_seasons
```

<h2>Total number of matches played in each season</h2>

```python
each_season=df['Season'].value_counts()
each_season = each_season.sort_index()
each_season
```

```python
each_season.plot(kind='bar', figsize=(10,7))

plt.title("Total number of matches played in each season")
plt.xlabel("Season")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

<h2>Maximum and minimum matches in a season</h2>

```python
each_season.max()
```

```python
each_season.min()
```

```python
each_season = each_season.sort_values(ascending=True)
plt.figure(figsize=(12,5))
plt.plot(each_season.index, each_season.values, marker='o')
plt.title("Maximum and minimum matches in a season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

<h2>Total number of matches won by each team</h2>

```python
team_win=df['WinningTeam'].value_counts()
team_win
```

```python
team_win = df['WinningTeam'].value_counts(ascending=True)
team_win.plot(kind='barh', figsize=(10,6))
plt.title("Total Wins by Each Team")
plt.xlabel("Number of Wins")
plt.ylabel("Team")
plt.show()
```

<h2>Total number of matches played in each standium</h2>


```python
each_stadium = df['Venue'].value_counts().shape
each_stadium
```

```python
each_stadium = df['Venue'].value_counts()
plt.figure(figsize=(14,6))
plt.bar(each_stadium.index, each_stadium.values)
plt.title("Total number of matches played in each standium")
plt.xlabel("Stadium (Venue)")
plt.ylabel("Number of Matches")
plt.xticks(rotation=90)
plt.show()
```

<h2>Has Toss-winning helped in Match-winning?</h2>

```python
df['TossHelped'] = df['TossWinner'] == df['WinningTeam']
df['TossHelped'].value_counts()
```

```python
toss_helped = df['TossHelped'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    toss_helped.values,
    labels=toss_helped.index.astype(str),
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Did Winning the Toss Help Win the Match?")
plt.show()
```

<h2>Top 15 player of the match winners</h2>
