```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
data.feature_names
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(precision_score(y_test,y_pred,average='macro'))
```

```python
print(recall_score(y_test,y_pred,average='macro'))
```

```python
print(f1_score(y_test,y_pred,average='macro'))
```

```python
#1>load data
#2>x,y
#3>split
#4>select model
import pandas as pd
df=pd.read_csv('StudentsPerformance.csv')
print(df.columns)
```

```python
df['average_score']=(df['math score']+df['reading score']+df['writing score'])/3
print(df['average_score'])
```

```python
df['result']=df['average_score'].apply(
    lambda x:'pass' if x>=50 else 'fail')
```

```python
print(df[['math score','reading score','writing score','result']])
```

```python
X=df[['math score','reading score','writing score']]
y=df['result']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
print(y_pred)
num_pass = sum(y_pred == 'pass')
num_fail = sum(y_pred == 'fail')
print("Number of students who passed:", num_pass)
print("Number of students who failed:", num_fail)
```
